# Sorting

## Descending

# Making changes to the data frame

## Add column

```python
#Add new column to the datframe
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

## Drop column

```python
df = df.drop(columns=['Total'])
df.head(5)
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(5)
```

# Saving data

## .csv format

```python
# index = False allows to remove the number of row
df.to_csv('Modified.csv', index = False)
```

## .xlsx format

```python
df.to_excel('Modified.xlsx', index = False)
```

## .txt format

```python
df.to_csv('Modified.txt', index = False)
```

# Filtering data

```python
# and condition with the selections
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
df.head(5)
```

```python
# or condition with the selections
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
df.head(5)
```

```python
#Other query
# or condition with the selections
new =df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison') & (df['HP']>70)]
#After this you can save the file
new.to_csv('new.csv', index = False)
df.head(5)
```

## Reset index

```python
df.drop(columns="#")
df.reset_index(drop=True)
```

# Conditional changes

```python
# Example 1
df.loc[df['Total'] > 500, ['Generation','Legendary']] = ['TEST 1', 'TEST 2']
df.head(5)
```

```python
# Example 2 with adding a column
df.loc[df['Total'] >500, ['Review']] = ['Great']
df.loc[df['Total'] <=500, ['Review']] = ['Bad']

df.head(5)
```

# Group by

```python
df = pd.read_csv('modified.csv')

df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python
df.groupby(['Type 1']).mean()
df.head(5)
```

```python
df.groupby(['Type 1']).sum()
df.head(5)
```
